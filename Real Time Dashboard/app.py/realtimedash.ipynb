#### EDA

```python
import warnings
warnings.filterwarnings("ignore")

from sklearn.preprocessing import LabelEncoder,OneHotEncoder 
le = LabelEncoder()
```

```python
categorical_columns = ['1st Team', '2nd Team', 'Stage','Top Scorer','Best Bowler']
numerical_columns = ['First Innings Score', 'Second Innings Score']

X = df[categorical_columns + numerical_columns]
y = df['Winners']
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Split data first
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=22)

# Initialize and fit the encoder
encoder = OneHotEncoder(drop='first', sparse_output=False, handle_unknown='ignore')

# Transform categorical columns
X_train_cat = encoder.fit_transform(X_train[categorical_columns])
X_test_cat = encoder.transform(X_test[categorical_columns])

# Combine with numerical columns
X_train = np.hstack([X_train_cat,X_train[numerical_columns].values])

X_test = np.hstack([X_test_cat,X_test[numerical_columns].values])


from sklearn.tree import DecisionTreeClassifier
# Train the model
classifier = DecisionTreeClassifier(random_state=22)
classifier.fit(X_train, y_train)

# Make predictions
y_pred = classifier.predict(X_test)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)* 100
print("Decision Tree Classifier")
print(f"Model Accuracy: {accuracy:.2f}")





from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
classifier2 = RandomForestClassifier(random_state=22)
classifier2.fit(X_train, y_train)
y_pred_train =classifier2.predict(X_train)
#accuracy = metrics.accuracy_score(y_test, y_pred)*100
print("----------------------------------------------------------------------")
print("Random Forest Classifier")
print("Test Accuracy: {:.2f}".format(accuracy))
#print(" accuracy(in %):", metrics.accuracy_score(y_test, y_pred)* 100)





#### logistic regression ####
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=1000)
model.fit(X_train,y_train)
y_pred=model.predict(X_test)
print("----------------------------------------------------------------------")
print("Logistic Regression Algorithm")
print("Test Accuracy: {:.2f}".format(accuracy))

#print("logistic regression accuracy(in %):", metrics.accuracy_score(y_test, y_pred)* 100

```

```python
from sklearn.naive_bayes import GaussianNB

gn=GaussianNB()
gn.fit(X_train,y_train)
y_pred=gn.predict(X_test)


from sklearn import metrics
print("Gaussian Naive Bayes")
print("gaussian naive model accuracy(in %):",metrics.accuracy_score(y_test,y_pred)*100)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
cm = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix using Seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
